## Question 1: 
Create a new column in the dataframe called "Profit", which is the Sale Price - Retail Price of each sneaker sold (i.e. each data point). Find the maximum, minimum, and average profit over all the data. 

```python
# Create a new column "Profit" by subtracting "Retail Price" from "Sale Price"
df['Profit'] = df['Sale Price'] - df['Retail Price']

# Calculate maximum, minimum, and average profit
max_profit = df['Profit'].max()
min_profit = df['Profit'].min()
avg_profit = df['Profit'].mean()

# Print the results
print("Maximum profit:", max_profit)
print("Minimum profit:", min_profit)
print("Average profit:", avg_profit)
```

What is the shoe name that gave the highest profit on a single sale?

```python
# Locate the row(s) where the profit is equal to the maximum profit
max_profit_rows = df[df['Profit'] == max_profit]

# Retrieve the shoe name(s) corresponding to the maximum profit
shoe_names_max_profit = max_profit_rows['Sneaker Type'].unique()

# Print the shoe name(s) that gave the highest profit on a single sale
print("Shoe name(s) with highest profit on a single sale:", shoe_names_max_profit)
```

What is the shoe name that gave the lowest profit on a single sale?

```python
# Locate the row(s) where the profit is equal to the minimum profit
min_profit_rows = df[df['Profit'] == min_profit]

# Retrieve the shoe name(s) corresponding to the minimum profit
shoe_names_min_profit = min_profit_rows['Sneaker Type'].unique()

# Print the shoe name(s) that gave the lowest profit on a single sale
print("Shoe name(s) with lowest profit on a single sale:", shoe_names_min_profit)
```

## Question 2

Group the data by Sneaker Type. Find the Sneaker Type that has the highest and lowest mean Profit. Then find that highest and lowest mean Profit value.

```python
# Group the data by "Sneaker Type" and calculate the mean profit for each group
mean_profit_by_type = df.groupby('Sneaker Type')['Profit'].mean()

# Find the Sneaker Type with the highest mean Profit
highest_mean_profit_type = mean_profit_by_type.idxmax()
highest_mean_profit_value = mean_profit_by_type.max()

# Find the Sneaker Type with the lowest mean Profit
lowest_mean_profit_type = mean_profit_by_type.idxmin()
lowest_mean_profit_value = mean_profit_by_type.min()

# Print the results
print("Sneaker Type with the highest mean Profit:", highest_mean_profit_type)
print("Highest mean Profit value:", highest_mean_profit_value)
print("Sneaker Type with the lowest mean Profit:", lowest_mean_profit_type)
print("Lowest mean Profit value:", lowest_mean_profit_value)
```

# Part 3

We provide you with the following function to model a set of dependent variables (x_columns) to a single independent variable (y_column) using linear regression. 

For example, you can relate Shoe Size and Days to Sale to Sale Price using the following:

do_ML_fit(df[["Shoe Size","Days to Sale"]], df["Sale Price"])

```python
def do_ML_fit(x_columns, y_column):
    lin_reg = lm.LinearRegression()
    lin_reg.fit(x_columns, y_column) 
    print('Coefficients:', lin_reg.coef_)
    print('Intercept:', lin_reg.intercept_)
    print(lin_reg.score(x_columns, y_column))
    print()
    return lin_reg
```

## Question 3

Which sneaker has the highest R^2 value for a single variable?

Which variable has the highest R^2 value overall?

If you look at the trend of Sale Price vs. Days to Sale for this shoe, beginning on day 15, what is the general trend?

Use the values from Question 1.

```python
# Dictionary to store the highest R2 value for each variable and its corresponding sneaker type
highest_r2_values = {"Shoe Size": {"value": 0, "sneaker": ""}, 
                     "Days to Sale": {"value": 0, "sneaker": ""}, 
                     "Retail Price": {"value": 0, "sneaker": ""}, 
                     "Count": {"value": 0, "sneaker": ""}}

# Iterate over each sneaker type
for t in sneaker_type_list:
    df_t = df[df["Sneaker Type"] == t]
    
    # Calculate R2 value for each variable
    r2_shoe_size = do_ML_fit(df_t[["Shoe Size"]], df_t["Sale Price"]).score(df_t[["Shoe Size"]], df_t["Sale Price"])
    r2_days_to_sale = do_ML_fit(df_t[["Days to Sale"]], df_t["Sale Price"]).score(df_t[["Days to Sale"]], df_t["Sale Price"])
    r2_retail_price = do_ML_fit(df_t[["Retail Price"]], df_t["Sale Price"]).score(df_t[["Retail Price"]], df_t["Sale Price"])
    r2_count = do_ML_fit(df_t[["Count"]], df_t["Sale Price"]).score(df_t[["Count"]], df_t["Sale Price"])
    
    # Update highest R2 values if higher value is found
    if r2_shoe_size > highest_r2_values["Shoe Size"]["value"]:
        highest_r2_values["Shoe Size"]["value"] = r2_shoe_size
        highest_r2_values["Shoe Size"]["sneaker"] = t
    
    if r2_days_to_sale > highest_r2_values["Days to Sale"]["value"]:
        highest_r2_values["Days to Sale"]["value"] = r2_days_to_sale
        highest_r2_values["Days to Sale"]["sneaker"] = t
    
    if r2_retail_price > highest_r2_values["Retail Price"]["value"]:
        highest_r2_values["Retail Price"]["value"] = r2_retail_price
        highest_r2_values["Retail Price"]["sneaker"] = t
    
    if r2_count > highest_r2_values["Count"]["value"]:
        highest_r2_values["Count"]["value"] = r2_count
        highest_r2_values["Count"]["sneaker"] = t

# Print the sneaker with the highest R2 value for each variable
for variable, values in highest_r2_values.items():
    print(f"Sneaker with the highest R2 value for {variable}: {values['sneaker']}")
    print(f"R2 value: {values['value']}")
    print()
```

## Question 8

Of the following 5 shoes, which sees the most significant rise in R^2 value when using Days to Sale and Count instead of just count as the variable(s) to the linear regression?

How much did the R^2 value increase in this case?  

```python
sneaker_list = ["Air-Jordan-1-Retro-High","Nike-Air-Max-97","Nike-Air-Presto","Nike-Blazer-Mid","Nike-Zoom-Fly"]
# Dictionary to store the R2 values for each shoe using "Count" as a variable
r2_values_count = {}

# Dictionary to store the increase in R2 value when adding "Days to Sale" as a variable
increase_r2_values = {}

# Iterate over each shoe
for shoe in sneaker_list:
    print(f"Calculating R2 values for {shoe}...")
    df_shoe = df[df["Sneaker Type"] == shoe]
    
    # Calculate R2 value using only "Count" as variable
    r2_count = do_ML_fit(df_shoe[["Count"]], df_shoe["Sale Price"]).score(df_shoe[["Count"]], df_shoe["Sale Price"])
    r2_values_count[shoe] = r2_count
    
    # Calculate R2 value using "Days to Sale" and "Count" as variables
    r2_days_to_sale_count = do_ML_fit(df_shoe[["Days to Sale", "Count"]], df_shoe["Sale Price"]).score(df_shoe[["Days to Sale", "Count"]], df_shoe["Sale Price"])
    
    # Calculate increase in R2 value when adding "Days to Sale" as a variable
    increase_r2_values[shoe] = r2_days_to_sale_count - r2_count
    print(increase_r2_values[shoe])

# Find the shoe with the most significant rise in R2 value
max_rise_shoe = max(increase_r2_values, key=increase_r2_values.get)

print(f"The shoe with the most significant rise in R2 value when using 'Days to Sale' and 'Count' as variables is: {max_rise_shoe}")
```

## Question 9

Rerun the linear regression model using all variables. Which sneaker results in the highest overall R^2 value?

```python
# Dictionary to store the R2 values for each sneaker type
r2_values = {}

# Iterate over each sneaker type
for shoe in df["Sneaker Type"].unique():
    df_shoe = df[df["Sneaker Type"] == shoe]
    
    # Selecting predictor variables
    x_columns = df_shoe[["Shoe Size", "Days to Sale", "Retail Price", "Count"]]
    
    # Calculate R2 value using all variables
    r2 = do_ML_fit(x_columns, df_shoe["Sale Price"]).score(x_columns, df_shoe["Sale Price"])
    
    # Store the R2 value for the sneaker type
    r2_values[shoe] = r2

# Find the sneaker type with the highest overall R2 value
max_r2_shoe = max(r2_values, key=r2_values.get)

print(f"The sneaker with the highest overall R^2 value is: {max_r2_shoe}")
```

### Question 2

Now let's take a look at what the data would look like for a small range of Sale Price: from 1960 to 2000. Run linear regression using degree = 1, 2, and 3 and for each case create a scatter plot of the slice of data together with the best fit curve found by the linear regression.   What is the $R^2$ value for degree = 1, 2, and 3? Round your answers to two decimal places.

```python
from sklearn.preprocessing import PolynomialFeatures

# Filter the DataFrame for the specified range of Sale Price
df_slice = df[(df["Sale Price"] >= 1960) & (df["Sale Price"] <= 1999)]

# Extract Sale Price and Profit
X = df_slice[["Sale Price"]]
y = df_slice["Profit"]

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data', color='blue')

# Sort X values
sorted_indices = np.argsort(X.values.ravel())
X_sorted = X.values[sorted_indices]
y_sorted = y.values[sorted_indices]

# Fit linear regression with degree 1
model_deg1 = LinearRegression()
model_deg1.fit(X_sorted, y_sorted)
y_pred_deg1 = model_deg1.predict(X_sorted)
plt.plot(X_sorted, y_pred_deg1, label='Degree 1', color='green')

# Fit polynomial regression with degree 2
poly_features_deg2 = PolynomialFeatures(degree=2)
X_poly_deg2 = poly_features_deg2.fit_transform(X_sorted)
model_deg2 = LinearRegression()
model_deg2.fit(X_poly_deg2, y_sorted)
y_pred_deg2 = model_deg2.predict(X_poly_deg2)
plt.plot(X_sorted, y_pred_deg2, label='Degree 2', color='orange')

# Fit polynomial regression with degree 3
poly_features_deg3 = PolynomialFeatures(degree=3)
X_poly_deg3 = poly_features_deg3.fit_transform(X_sorted)
model_deg3 = LinearRegression()
model_deg3.fit(X_poly_deg3, y_sorted)
y_pred_deg3 = model_deg3.predict(X_poly_deg3)
plt.plot(X_sorted, y_pred_deg3, label='Degree 3', color='red')

plt.xlabel("Sale Price")
plt.ylabel("Profit")
plt.title("Scatter Plot with Best Fit Curves")
plt.legend()
plt.grid(True)
plt.show()
```

For each of the plots which show the best fit curve for degree = 1, 2, and 3, determine which is an example of overfitting to the overall data. Select all degrees that are an example of overfitting.
